```python
import random
import datetime
import csv
from prettytable import PrettyTable, ALL
import os
import re



# Define a list to store the questions
questions = []

               
# Function to add a new question to the list
def add_question(question_text, responses, correct_response, topic_tags):
    question = [question_text, responses, correct_response, topic_tags]
    questions.append(question)

    
# Function to save the list of questions to a .csv file
def save_questions(questions):
    try:
        # check if the file exists
        if not os.path.exists('questions.csv'):
            # create the file and write the header row
            with open('questions.csv', 'w', newline='') as file:
                writer = csv.writer(file)
                writer.writerow(['Question', 'Responses', 'Correct Response', 'Topic Tags'])

        # open the existing CSV file in "append" mode
        with open('questions.csv', 'a', newline='') as file:
            # create a CSV writer object
            writer = csv.writer(file)
            # write each question to the CSV file
            for question in questions:
                writer.writerow(question)

    except Exception as e:
        print(f"An error occurred while saving the questions: {e}")


# Function to load the list of questions from a .csv file
def load_questions():
    try:
        with open('questions.csv', 'r', newline='') as file:
            reader = csv.reader(file)
            next(reader)  # Skip the header row
            for row in reader:
                question_text, responses, correct_response, topic_tags = row
                add_question(question_text, responses, correct_response, topic_tags)
    except FileNotFoundError:
        print("The file 'questions.csv' does not exist.")
    except Exception as e:
        print(f"An error occurred while loading the questions: {e}")

    
    
    
    
    
    
    

        
# Function to show the table of stored questions.
def show_table_of_questions():

    # Open the CSV file and read the contents
    with open("questions.csv", "r") as file:
        reader = csv.reader(file)
        data = list(reader)

    # Create a PrettyTable object and add the data
    table = PrettyTable()
    table.field_names = data[0]
    for row in data[1:]:
        table.add_row(row)

    # Add a horizontal line between each row
    table.hrules = ALL


    # Set the maximum width of the table manually
    table.max_width = 26


    # Display the table
    print(table)

        

#display a question
def display_question(question_number, question):
    
    print(f"{question_number}) {question[0]}")
    possible_responses_list = question[1].split(';')    

    i = 0
    for posible_response in possible_responses_list:
        print(f"{chr(65+i)}) {posible_response}")
        i += 1

    
def check_the_answer(question, selected_answer):
    # Define a list to store the choices
    valid_choice = ["A", "B", "C", "D"]
    possible_responses_list = question[1].split(';') 
    selected_answer_index = valid_choice.index(selected_answer)
    if selected_answer not in valid_choice:
        print("The answer you entered does not exist. Please select your answer from the list")
    elif possible_responses_list[selected_answer_index] == question[2]:
        return(True)
    else:
        return(False)
                   
    
def whats_the_correct_answer(question):
    valid_choice = ["A", "B", "C", "D"]
    possible_responses_list = question[1].split(';') 
    correct_response_index = possible_responses_list.index(question[2])
    valid_choice[correct_response_index]
    print(f"The correct answer is: {valid_choice[correct_response_index]}")
       

# Calculate the average time taken to answer all questions
def average_time(start_list, end_list):
    time_taken = []
    for i in range(len(end_list)):
        difference = end_list[i] - start_list[i]
        time_taken.append(difference.total_seconds())  # convert to seconds
        average = sum(time_taken) / len(time_taken)
    return average
    


    
    
    
    
    

# Define study_session function
def study_session():
    
    question_number = 0
    number_of_correct_answers = 0
    total_number_of_questions = 0
    number_of_question = 0
    starttime = []
    endtime = []
    
    continue_or_not = "Yes"
    
    while continue_or_not.upper() in ["Y", "YES"]:

        #load the questions
        load_questions()
        
        if number_of_correct_answers == len(questions):
            # print all the questions has been answered
            print("\nAll the questions has been answered")
            break
        
        while questions:
            
            #select a question randomly
            random_question = random.choice(questions)


            #display a question
            print("\n")
            question_number += 1
            display_question(question_number, random_question)
            print("\n")


            # start the timer
            start_time = datetime.datetime.now()
            starttime.append(start_time)

            #recieve the user's response
            valid_choices = ['A', 'B', 'C', 'D']
            selected_answer = input("Please choose the correct answer ")
            capital_selected_answer = selected_answer.upper()
            while capital_selected_answer not in valid_choices:
                print("Invalid choice. Please choose one of the following:", valid_choices)
                selected_answer = input("Please choose the correct answer ")
                capital_selected_answer = selected_answer.upper()


            # stop the timer
            end_time = datetime.datetime.now()
            endtime.append(end_time)


            #check the answer
            if check_the_answer(random_question, capital_selected_answer) == True:
                print("Thats correct. Good job ")
                number_of_correct_answers += 1
                total_number_of_questions += 1
                questions.remove(random_question)
               
            else:
                print("You have chosen the wrong answer ")
                total_number_of_questions += 1
                whats_the_correct_answer(random_question)
                
                


            #check if the user wants to answer another question or not
            continue_or_not = input("Would you like to continue? Y or N ")
            if continue_or_not.lower() == 'n':
                break
                
     
    #print the number of correct answers
    print(f"\nYou have answered {number_of_correct_answers} of {total_number_of_questions} questions correctly. ")
    
    #print the average time to answer the questions
    average = average_time(starttime, endtime)
    print(f"Average time taken to answer all questions: {average} seconds")
    
    #ask user if they whant to create question or not
    want_to_exit = input(f"Do you want to create any questions or exit?\nEnter 1 to create a question or any other key to exit: ")
    if want_to_exit == "1":
        question_creation()
    else:
        exit()

  



    
    
# Define question_creation function
def question_creation():

    ans = "Y"
    while ans.upper() in ["Y", "YES"]:
        qt = input("Please enter the question text:\n")

        pr = input("Please enter the list of possible responses separated by ';':\n")
        pattern = r'^[a-zA-Z0-9]+(?:\s+[a-zA-Z0-9]+)*;[a-zA-Z0-9]+(?:\s+[a-zA-Z0-9]+)*;[a-zA-Z0-9]+(?:\s+[a-zA-Z0-9]+)*;[a-zA-Z0-9]+(?:\s+[a-zA-Z0-9]+)*$'

        while True:
            while not re.match(pattern, pr):
                pr = input("Please enter 4 responses separated by ';'\nPlease ensure there are no extra spaces between the options.\nOnly ';' is allowed as a delimiter between the options.\n")

            valid_pr = pr.split(';')
            if len(set(valid_pr)) == len(valid_pr):
                break
            else:
                pr = input("Some of the answers are repetitive\nPlease enter 4 distinct possible responses:\n")

        cr = input("Designate which response is correct:\n")
        while cr not in pr.split(";"):
            cr = input("The answer you chosen does not exist in the possible responses.\nPlease choose the correct answer among the possible responses!\n")

        tt = input("Please enter the list of topic tags separated by ';':\n")
        pattern = r'^[a-zA-Z0-9]+(?:\s+[a-zA-Z0-9]+)*(?:;[a-zA-Z0-9]+(?:\s+[a-zA-Z0-9]+)*)*(?<!\s|;) *$'
        while not re.match(pattern, tt):
            tt = input("Please follow the instruction.\nPlease enter the list of topic tags separated by ';'\nPlease ensure there are no extra spaces between the options.\nOnly ';' is allowed as a delimiter between the options.\n")

        # check if the new question already exists
        exists = False
        for q in questions:
            if qt == q[0] and pr == q[1] and tt == q[3]:
                exists = True
                break

        if exists:
            print("This question already exists.")
        else:
            # add the new question to the list of questions
            add_question(qt, pr, cr, tt)

        ans = input("Do you want to add another question: Y or N ")
        while ans.upper() not in ["N", "NO", "Y", "YES"]:
            ans = input("Please enter a correct response. 'Y' for Yes or 'N' for No ")
        if ans.upper() in ["N", "NO"]:
            want_to_quit = input("Do you want to begin a study session or exit?\nEnter 1 to begin a study session or any other key to exit: ")
            if want_to_quit == "1":
                save_questions(questions)
                show_table_of_questions()
                study_session()
            else:
                save_questions(questions)
                show_table_of_questions()
                exit()

    # save the list of questions to the .csv file
    save_questions(questions)

    # show the table of stored questions
    show_table_of_questions()

    
    
        
    

# Welcome messages
print("Welcome to the application.")
print("Please select an option:")
print("1. Begin a study session")
print("2. Add a new question to the list")
print("3. exit\n")

# Prompt for input
user_input = ""
while user_input not in ["1", "2", "3"]:
    user_input = input("Please enter your choice (1, 2 or 3): ")

    if user_input == "1":
        study_session()
     
    elif user_input == "2":
        question_creation()
    
    elif user_input == "3":
        exit()
            
    else:
        print("\nInvalid input.")
      
```
